<span style="color: blue">1. Write a Python program to simulate the following scenarios:  
  a. Tossing a coin 10,000 times and calculating the experimental probability of heads and tails.  
  b. Rolling two dice and computing the probability of getting a sum of 7.  
  Steps  
      a. Use Python's random module for simulations.  
      b. Implement loops for repeated trials.  
      c. Track outcomes and compute probabilities. 
</span> 

```python
import random

def coin_tossed(num_of_tosses=10000):
    heads = 0    
    tails = 0
    
    for _ in range(num_of_tosses):  
        if random.choice(["heads", "tails"]) == "heads":  # Choosing randomly between 'heads' and 'tails'
            heads += 1      # Counting a head when it occurs
        else:
            tails += 1      # Counting a tails when it occurs
            
    print(f"Coin simulations: {num_of_tosses}")   
    print(f"Probability of heads: {heads}/{num_of_tosses} = {heads / num_of_tosses:.2%}")
    print(f"Probability of tails: {tails}/{num_of_tosses} = {tails / num_of_tosses:.2%}")
    
# Calling the function to start tossing coins    
coin_tossed() 
```

<span style="color: purple;  font-weight: bold;">I wrote a program to toss a coin 10,000 times. I used random choice to get either heads or tails. I counted how many times heads came and how many times tails came. Then I found the probability of both by dividing their count by 10,000. This helped me understand how often heads or tails comes when we toss a coin many times.<span>

<span style="color: blue">2. Write a function to estimate the probability of getting at least one "6" in 10 rolls of a fair die.  
  Steps  
      a. Simulate rolling a die 10 times using a loop.  
      b. Track trials where at least one "6" occurs.  
      c. Calculate the proportion of successful trials.
.</span>  

```python
import random

def dice_rolled(no_trials):
    count_of_6 = 0
    
    for _ in range(no_trials):
        rolls = [random.randint(1, 6) for _ in range(10)]  # Rolling the die 10 times and saving results 
        if 6 in rolls:       # Checking if any roll is 6
            count_of_6 += 1  # Counting when 6 appears in those 10 rolls
            
    return count_of_6 / no_trials  # Calculating probability of getting at least one 6 in 10 rolls

# Calling function with 10,000 trials to estimate probability
probability = dice_rolled(10000)
print(f'Probability: {probability:.4f}')
    
```

<span style="color: purple;  font-weight: bold;">I made a program to roll a die 10 times in each trial and repeated this 10,000 times. In each trial, I checked if at least one “6” appeared. I counted how many trials had a “6” and then divided that count by the total trials to get the probability. This tells how likely it is to get at least one “6” in 10 rolls.<span>

<span style="color: green; font-weight: bold;">Conditional Probability and Baye's Theorem </span>

<span style="color: blue">3. A bag contains 5 red, 7 green, and 8 blue balls. A ball is drawn randomly, its color noted, and it is put back into the bag. If this process is repeated 1000 times, write a Python program to estimate:  
  a. The probability of drawing a red ball given that the previous ball was blue.  
  b. Verify Bayes' theorem with the simulation results.  
Steps  
    a. Use random sampling to simulate the process.  
    b. Compute conditional probabilities directly from the data.
</span> 

```python
import random

def simulate_draws(num_trials=1000):
    colors = ['Red'] * 5 + ['Green'] * 7 + ['Blue'] * 8  
    draws = [random.choice(colors) for _ in range(num_trials)]  
    
    # Count of each occurrences
    count_blue_before = 0  
    count_red_after_blue = 0  
    
    for i in range(1, num_trials):
        if draws[i - 1] == 'Blue':  
            count_blue_before += 1  
            if draws[i] == 'Red':
                count_red_after_blue += 1  

    # Estimated P(R | B)
    prob_red_given_blue = count_red_after_blue / count_blue_before if count_blue_before > 0 else 0 #conditional prob

    # Computing theoretical probabilities for Bayes theorem verification
    P_R = 5 / 20 
    P_B = 8 / 20  
    P_B_given_R = P_B  

    # Bayes' Theorem Calculation
    bayes_result = (P_B_given_R * P_R) / P_B  

    return prob_red_given_blue, bayes_result


simulated_prob, bayes_prob = simulate_draws(1000)

# Printing results
print(f"Simulated P(Red | Blue): {simulated_prob:.4f}")
print(f"Bayes' Theorem P(Red | Blue): {bayes_prob:.4f}")
```

<span style="color: purple; font-weight: bold;">
I created a program that simulates picking balls from a bag 1000 times. The bag has red, green, and blue balls. I noted each color picked and then checked how often a red ball came after a blue one. From that, I calculated the probability of getting red given that blue came before.  
Then, I used Bayes’ Theorem to calculate the same probability in theory and compared both results. This helps to check if the simulation matches the math.
</span>

<div style="color: blue; white-space: pre-wrap;">
4. Generate a sample of size 1000 from a discrete random variable with the following distribution:  
 - P(X=1) = 0.25  
 - P(X=2) = 0.35  
 - P(X=3) = 0.4  
Compute the empirical mean, variance, and standard deviation of the sample.  
Steps  
a. Use numpy.random.choice() to generate the sample.  
b. Use numpy methods to calculate mean, variance, and standard deviation.
</div>


```python
import numpy as np

# Values and their probabilities
values = [1, 2, 3]
probabilities = [0.25, 0.35, 0.4]

# Generating a sample of size 1000
sample = np.random.choice(values, size=1000, p=probabilities)

# Computing empirical mean, variance, and standard deviation
mean_empirical = np.mean(sample)
variance_empirical = np.var(sample)
std_dev_empirical = np.std(sample)

# Displaying results
print(f"Empirical Mean: {mean_empirical:.4f}")
print(f"Empirical Variance: {variance_empirical:.4f}")
print(f"Empirical Standard Deviation: {std_dev_empirical:.4f}")
```

<span style="color: purple; font-weight: bold;">
I wrote a program where I randomly picked numbers from the list [1, 2, 3], based on given probabilities:  
25% chance for 1, 35% for 2, and 40% for 3. I repeated this 1000 times.  

Then I calculated the average (mean), how much the numbers vary (variance), and how spread out they are (standard deviation).  
This shows how the numbers behave when chosen many times.
</span>

<div style="color: blue; white-space: pre-wrap;">5. Simulate 2000 random samples from an exponential distribution with a mean of 5. Visualize the distribution using:  
  a. A histogram.  
  b. A probability density function (PDF) overlay.  
  Steps  
      a. Use numpy.random.exponential().  
      b. Use matplotlib to create visualizations.
<div>

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import expon
```

```python
mean_val = 5
samples = np.random.exponential(scale=mean_val, size=2000)
```

```python
# Plotting histogram 

plt.figure(figsize=(8, 5))
sns.histplot(samples, bins=30, stat="density", color='blue', alpha=0.6, label="Histogram")
```

```python
# plot
x_vals = np.linspace(0, max(samples), 1000)
pdf_vals = expon.pdf(x_vals, scale=mean_val)
plt.plot(x_vals, pdf_vals, 'r', label="PDF")

plt.title("Exponential Distribution (Mean = 5)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

<span style="color: purple; font-weight: bold;">
I created 2000 random numbers from an exponential distribution where the average value is 5.  
This means most numbers are small but some can be very big.  

We use this to model things like waiting times between events.

<div style="color: blue; white-space: pre-wrap;">6. Simulate the Central Limit Theorem by following these steps  
  a. Generate 10,000 random numbers from a uniform distribution.  
  b. Draw 1000 samples of size n = 30.  
  c. Calculate and visualize the distribution of sample means.  
  Steps  
      a. Use numpy.random.uniform().  
      b. Plot both the uniform distribution and the sample mean distribution for comparison.
<div>

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generating 10,000 random numbers from a uniform distribution (0,1)
uniform_data = np.random.uniform(0, 1, 10000)
```

```python
# Drew 1000 samples of size n=30 and computed sample means
n_samples = 1000 
sample_size = 30  
sample_means = [np.mean(np.random.choice(uniform_data, sample_size, replace=True)) for _ in range(n_samples)]
```

```python
# Plotting the uniform distribution
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
sns.histplot(uniform_data, bins=30, kde=True, color='blue', alpha=0.6)
plt.title("Original Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Plotting the distribution of sample means
plt.subplot(1, 2, 2)
sns.histplot(sample_means, bins=30, kde=True, color='green', alpha=0.6)
plt.title("Distribution of Sample Means (n=30)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

<span style="color: purple; font-weight: bold;">
I made 10,000 random numbers from a uniform distribution (all values equally likely between 0 and 1).  
Then, I took 1000 samples, each with 30 numbers, and found the average of each sample.  
Finally, I drew two graphs: one for the original numbers and one for the averages.  
This shows that even if the original data is flat, the averages make a bell-shaped curve, proving the Central Limit Theorem.  
</span>
