1. Create a simple line plot for the following data:<br><br>
x = [1, 2, 3, 4, 5]<br>
y = [10, 15, 25, 30, 50]<br><br>
a. Plot the data.<br>
b. Customize the plot by adding a title, axis labels, and a grid.
</div>


```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

# Plot 
plt.plot(x, y, marker='o', linestyle='-', color='b', label="Line Plot")

# Customization
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True,linestyle='--') 
plt.legend()  

# Show the plot
plt.show()
```

<div style="background-color:#00008B; color:white; padding:10px; margin:10px 0;"> This code creates a simple line plot using two lists: x and y. It draws points connected by a blue line. The plot has a title and labels for both axes. It also shows grid lines with dashed style to make reading easier. A legend is added to explain the line. </div>

<div style="background-color:#4B0082; color:white; padding:10px; margin:10px 0;">
2. Create a bar graph to represent the marks scored by students in a subject:<br><br>
students = ['John', 'Jane', 'Alice', 'Bob']<br>
marks = [75, 85, 60, 90]<br><br>
a. Plot the data as a bar graph.<br>
b. Customize the colors and add a title.
</div>


```python
import matplotlib.pyplot as plt

# Data
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

# Plot 
plt.bar(students, marks, color=['lightblue', 'lightgreen', 'yellow', 'purple'])

# Customization
plt.title("Marks Scored by Students")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.ylim(0, 100)  
plt.grid(axis='y', linestyle='--',color = 'green' ,alpha=0.9)  

# Show the plot
plt.show()
```

<div style="background-color:#00008B; color:white; padding:10px; margin:10px 0;"> This code makes a bar graph to show marks of four students: John, Jane, Alice, and Bob. Each student’s mark is shown as a bar with different colors. The graph has a title and labels on both sides to explain what is shown. The y-axis goes from 0 to 100, and there are green lines to help see the values better. </div>

3. Create a pie chart to represent the percentage distribution of a company’s revenue from different regions:<br><br>
regions = ['North America', 'Europe', 'Asia', 'Others']<br>
revenue = [45, 25, 20, 10]<br><br>
a. Create a pie chart with the region names as labels.<br>
b. Highlight the region with the highest revenue.
</div>


```python
import matplotlib.pyplot as plt

# Data
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

# Highlighting the region with the highest revenue (North America)
explode = [0.02, 0, 0, 0]

# Plot 
plt.figure(figsize=(7, 7))  
plt.pie(revenue, labels=regions, autopct='%1.1f%%', colors=['blue', 'green', 'red', 'purple'],
        explode=explode, startangle=80)

# Title
plt.title("Revenue Distribution by Region")

# Show plot
plt.show()
```

<div style="background-color:#00008B; color:white; padding:10px; margin:10px 0;"> This code draws a pie chart to show how much money comes from different places: North America, Europe, Asia, and Others. North America makes the most money, so its slice is pulled out a little to stand out. Each region has a different color. The chart also shows percentages on each slice and has a clear title. </div>

4. Generate a histogram to show the frequency distribution of randomly generated integers between 1 and 100 (sample size = 1000).
</div>


```python
import matplotlib.pyplot as plt
import numpy as np

# Generates 1000 random integers between 1 and 100
data = np.random.randint(1, 101, size=1000)

# Plot histogram
plt.figure(figsize=(8, 5))  
plt.hist(data, bins=10, color='g', edgecolor='black', alpha=0.7)

# Customization
plt.title("Frequency Distribution of Random Integers (1-100)")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--',color = 'b', alpha=0.7)

# Show plot
plt.show()
```

<div style="background-color:#00008B; color:white; padding:10px; margin:10px 0;"> In this code, I created 1000 random whole numbers between 1 and 100 using NumPy. Then, I made a histogram to show how often numbers appear in different ranges (bins). I added labels for the title, x-axis, and y-axis to explain the chart. I also put a blue dotted grid on the y-axis to make the chart easier to read. Finally, I displayed the histogram using matplotlib. </div>
